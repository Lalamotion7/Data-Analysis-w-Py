```python
#Importanndo as bibliotecas necessárias
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

#carregando as credenciais
load_dotenv()

#Criando a conexão com o BD
engine = create_engine(f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASS')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}")

#Query SQL para extração dos dados
query_extracao = "SELECT * FROM help_desk_tickets"

#Armazenando os dados em um DataFrame
df_help_desk = pd.read_sql(query_extracao, engine)

#Exibindo as primeiras linhas do DataFrame
print(df_help_desk.head())
```

```python
# Informações sobre o DataFrame
df_help_desk.info()
```

```python
#Estatísticas descritivas do DF
df_help_desk.describe(include='all')
```

```python
# Verificando os valores ausentes
df_help_desk.isnull().sum()
```

```python
#Verificando qual é o status dos chamados que não possuem data de fechamento
filtro_vazios = df_help_desk[df_help_desk['data_fechamento'].isnull()]
print(filtro_vazios['status'].value_counts())
```
